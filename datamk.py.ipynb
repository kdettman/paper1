```python
import numpy as np

#reading in velocity, color, shape, and Hubble residual information
f = open('velcolsigma2.txt','r')
fl = f.readlines()[1:]

name = [x.split()[0] for x in fl]
col  = [float(x.split()[1]) for x in fl]
colerr = [float(x.split()[2]) for x in fl]
vel  = [float(x.split()[3]) for x in fl]
velerr = [float(x.split()[4]) for x in fl]
x1 = [float(x.split()[5]) for x in fl]
x1err = [float(x.split()[6]) for x in fl]
mures = [float(x.split()[7]) for x in fl]
muerr = [float(x.split()[8]) for x in fl]
vel_up = [x.split()[9] for x in fl]

#reading in redshift and phase information
#this file also contains path information on the spectra
fi = open('foundInstr.txt','r')
fil = fi.readlines()[16:]

name_i = [x.split()[0] for x in fil]
mjdobs = [float(x.split()[5]) for x in fil]
pkmjd = [float(x.split()[6]) for x in fil]
z = [float(x.split()[3]) for x in fil]

#ATEL velocities
fu = open('unmeasuredVel.txt','r')
ful = fu.readlines()[1:]
name_un = [x.split()[0] for x in ful]
vel_un = [float(x.split()[1])/-1000.0 for x in ful]
velerr_un = np.mean(velerr)*np.ones(len(name_un)) #ATEL velocities have no error so I am giving them the average error of the measured sample
col_un = [x.split()[2] for x in ful]
colerr_un = [x.split()[3] for x in ful]
x1_un = [x.split()[4] for x in ful]
x1err_un = [x.split()[5] for x in ful]
mures_un = [x.split()[6] for x in ful]
muerr_un = [x.split()[7] for x in ful]
mjdobs_un = [float(x.split()[8]) for x in ful]
pkmjd_un = [x.split()[9] for x in ful]
z_un = [x.split()[9] for x in ful]

index = []
phase = []
mask = np.ones(len(name), np.bool)
for i in range(len(name)):
    phase_tmp = (mjdobs[i]-pkmjd[i])/(1.+z[i])
    phase.append(phase_tmp)
    if ('#' in name[i] or abs(vel[i])>50.0 or vel[i]>0 or x1[i]>3.0 or col[i]>0.319 or phase_tmp<-6 or phase_tmp>10):
        #making cuts to the full dataset according to color, shape, and non-sensical velocities or specially marked poor spectra (#)
        index.append(i)
mask[index] = False
name_n = np.asarray(name)[mask]
col_n = np.asarray(col)[mask].astype('float')
colerr_n = np.asarray(colerr)[mask].astype('float')
vel_n = np.asarray(vel)[mask].astype('float')
velerr_n = np.asarray(velerr)[mask].astype('float')
x1_n = np.asarray(x1)[mask].astype('float')
x1err_n = np.asarray(x1err)[mask].astype('float')
mures_n = np.asarray(mures)[mask].astype('float')
muerr_n = np.asarray(muerr)[mask].astype('float')
phase_n = np.asarray(phase)[mask].astype('float')

mask_o = np.invert(mask)
name_o = np.asarray(name)[mask_o]
col_o = np.asarray(col)[mask_o].astype('float')
colerr_o = np.asarray(colerr)[mask_o].astype('float')
vel_o = np.asarray(vel)[mask_o].astype('float')
velerr_o = np.asarray(velerr)[mask_o].astype('float')
x1_o = np.asarray(x1)[mask_o].astype('float')
x1err_o = np.asarray(x1err)[mask_o].astype('float')
mures_o = np.asarray(mures)[mask_o].astype('float')
muerr_o = np.asarray(muerr)[mask_o].astype('float')
phase_o = np.asarray(phase)[mask_o].astype('float')

index2 = []
phase_un = []
mask2 = np.ones(len(name_un),np.bool)
for i in range(len(name_un)):
    if (col_un[i] != 'none' and pkmjd[i] != 'none'):
        phase_tmp = (float(mjdobs_un[i])-float(pkmjd_un[i]))/(1.+float(z_un[i]))
        phase_un.append(phase_tmp)
        if ('#' in name[i] or abs(vel[i])>50.0 or vel[i]>0 or x1[i]>3.0 or col[i]>0.319 or phase_tmp<-6 or phase_tmp>10):
            index2.append(i)
    else:
        index2.append(i)
        phase_un.append('none')
            
mask2[index2] = False
name_unn = np.asarray(name_un)[mask2]
vel_unn = np.asarray(vel_un)[mask2].astype('float')
velerr_unn = np.asarray(velerr_un)[mask2].astype('float')
col_unn = np.asarray(col_un)[mask2].astype('float')
colerr_unn = np.asarray(colerr_un)[mask2].astype('float')
x1_unn = np.asarray(x1_un)[mask2].astype('float')
x1err_unn = np.asarray(x1err_un)[mask2].astype('float')
mures_unn = np.asarray(mures_un)[mask2].astype('float')
muerr_unn = np.asarray(muerr_un)[mask2].astype('float')
phase_unn = np.asarray(phase_un)[mask2].astype('float')

#saving data for easy future use
np.savez('data_full.npz',name=name,col=col,colerr=colerr,vel=vel,velerr=velerr,x1=x1,x1err=x1err,mures=mures,muerr=muerr,phase=phase)
np.savez('data_wcuts.npz',name=name_n,col=col_n,colerr=colerr_n,vel=vel_n,velerr=velerr_n,x1=x1_n,x1err=x1err_n,mures=mures_n,muerr=muerr_n,phase=phase_n)
np.savez('data_cut.npz',name=name_o,col=col_o,colerr=colerr_o,vel=vel_o,velerr=velerr_o,x1=x1_o,x1err=x1err_o,mures=mures_o,muerr=muerr_o,phase=phase_o)
np.savez('data_un.npz',name=name_unn,col=col_unn,colerr=colerr_unn,vel=vel_unn,velerr=velerr_unn,x1=x1_unn,x1err=x1err_unn,mures=mures_unn,muerr=muerr_unn,phase=phase_unn)

f.close()
fi.close()
fu.close()
```
